```python
def sum_index(pdf):
    t=0
    k=pdf
    for i in range(4):
        t+=k[1][i][0]
        
    return t

    
```

```python
def indexpoz(pdf):
    t=0
    k=pdf
    t+=k[1][0][0]
    t+=k[1][0][1]
    
        
    return t
```

```python
def bubbleSort(prd):
    
  # loop to access each array element
  for i in range(len(prd)):

    # loop to compare array elements
    for j in range(0, len(prd) - i - 1):

      # compare two adjacent elements
      # change > to < to sort in descending order
        if sum_index(prd[j])> sum_index(prd[j+1]):

        # swapping elements if elements
        # are not in the intended order
            temp = prd[j]
            prd[j] = prd[j+1]
            prd[j+1] = temp
```

```python
def list_revorse(list):
    list.reverse()
    return list
```

```python
bubbleSort(prd)
list_revorse(prd)
```
